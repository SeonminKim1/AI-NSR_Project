## OP0 비교
- 1. Predict 한 결과 Union 및 실제 정답 데이터와 성능평가

```python
import os
import numpy as np
import pandas as pd

# file_list op0
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# 정답 데이터
file_list_op0 = os.listdir('data/binutils_answer/0_answer')
csv_op0 = []
for f in file_list_op0:
    csv = pd.read_csv('data/binutils_answer/0_answer/'+f, index_col=0)
    csv_op0.append(csv)
#    print(f, csv.shape, csv['label'].value_counts())
    
# 선민, 두혁꺼
file_list_op0_predict_seon = os.listdir('data/predict_result/bin_op0')
file_list_op0_predict_jang = os.listdir('data/predict_result_jang/bin_op0')

final_predict_csv_op0 = []

# 선민, 두혁꺼
for f1, f2 in zip(file_list_op0_predict_seon, file_list_op0_predict_jang):
    csv_predict_seon = pd.read_csv('data/predict_result/bin_op0/' + f1, index_col=0).loc[1:] # 선민
    csv_predict_jang = pd.read_csv('data/predict_result_jang/bin_op0/'+f2, index_col=0).loc[1:] # 두혁
    union_op0 = pd.concat([csv_predict_seon, csv_predict_jang['predict']],axis=1)
    union_op0['label'] = union_op0['label'].astype(int) 
    union_op0['predict'] = union_op0['predict'].astype(int)
    union_op0['final_label'] = union_op0['label'] | union_op0['predict']
    final_predict_csv_op0.append(union_op0)
    
# 최종 TEST
count=0
for k, k2 in zip(final_predict_csv_op0, csv_op0):
    pred = k['final_label'].values
    y_test = k2['label'].values

    k_accuracy = np.round(float(accuracy_score(pred, y_test)),4)
    k_recall =  np.round(float(recall_score(pred, y_test)),4)
    k_precision = np.round(float(precision_score(pred, y_test)),4)
    k_f1_score = np.round(float(f1_score(pred, y_test)),4)
    print('op0 ',count+1, ',' , file_list_op0[count], ': ', k_accuracy, k_recall, k_precision, k_f1_score)
    count+=1
```
